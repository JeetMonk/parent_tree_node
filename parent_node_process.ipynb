```python
import datetime
import pytz
import pandas as pd
import glob
import os
import numpy as np
```

```python
def printlog(message, levelname):
  __tz = pytz.timezone('Australia/Sydney')
  __ct = datetime.datetime.now(tz=__tz)
  __current_time = __ct.strftime("%d/%m/%Y %H:%M:%S")
  if levelname.lower() == "i":
    print("[%s] - %s - %s" % (__current_time, "INFO", message))
  elif levelname.lower() == "e":
    print("[%s] - %s - %s" % (__current_time, "ERROR", message))
  elif levelname.lower() == "d":
    print("[%s] - %s - %s" % (__current_time, "DEBUG", message))
  elif levelname.lower() == "w":
    print("[%s] - %s - %s" % (__current_time, "WARNING", message))
  else:
    print("[%s] - %s - %s" % (__current_time, "UNKNOWN", message))
```

```python
def find_parent_branch_id_list(branchid, parent_branch_id_list=None):
  if parent_branch_id_list is None:
    parent_branch_id_list = []
  parent_brach_id = dfBranch[dfBranch['branchid']==branchid]['parentbranchid'].iloc[0]
  if parent_brach_id != '0':
    parent_branch_id_list.append(parent_brach_id)
    return find_parent_branch_id_list(parent_brach_id, parent_branch_id_list)
  else:
    return parent_branch_id_list
```

```python
printlog("Starts to load  Branch data...","i")
printlog("Accoring to the file name find the latest file from input path: C:\\Users\\jeetk\\Desktop\\python\\data\\", "i")
BranchPath = "C:\\Users\\jeetk\\Desktop\\python\\data\\"
BranchFilePath = glob.glob(BranchPath + "/*.dat")
BranchSortedFileList = sorted(BranchFilePath, reverse=True)
printlog("The picked Branch file is: " + BranchSortedFileList[0],"i")

dfBranch = pd.read_csv(BranchSortedFileList[0], delimiter='|', header=0, converters={'branchid':str, 'contactdetailid':str, 'parentbranchid':str, 'name':str, 'description':str})
dfBranch.count()
```

```python
printlog("Starts to generate all parent branch ids for each branch id...","i")
dfBranch_pd = pd.DataFrame(columns=['branchid', 'parentbranchid'])
for branchid in dfBranch['branchid'].values:
  parent_branch_id_list = find_parent_branch_id_list(branchid)
  for parent_branch_id in parent_branch_id_list:
    dfBranch_pd = dfBranch_pd.append({'branchid': branchid, 'parentbranchid': parent_branch_id}, ignore_index=True)
printlog("Parent branch ids generation completed","i")
dfBranch_pd.count()
```
